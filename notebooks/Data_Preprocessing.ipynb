### After merging the above datasets the distribution of the label column was analyzed to understand the types of network traffic present in the dataset.
The above are the counts were obtained

### Exp: The Label column represents the type of network traffic captures in the each flow record. each label corresponds to a specific type of network behaviour:
- Bengin -> Normal Network Traffic(non-malicious)
- Syn -> SYN Flood Attack
- DrDoS_UDP -> UDP-Based Distributed Denial of Service Attack
- UDP-Lag -> UDP Lag Attack Pattern
- WebDDoS -> Web based Distributed Denial Of Service Attack

### Observations
- The dataset is dominated by attack traffic, which is expected since each dataset file represents a specific DDoS attack scenario.

- The BENIGN traffic represents normal network activity, but its count is significantly smaller compared to attack samples.

- The presence of multiple attack types helps the model learn different patterns of malicious network behavior.

- This dataset structure enables the development of a machine learning model capable of distinguishing between normal traffic and DDoS attack traffic.

### Null & Infinity Value handling

```python
# --- Handling the Missing Values ---

data['Source Port'] = data['Source Port'].fillna(data['Source Port'].median())
print("Source Port Missing Values: ", data['Source Port'].isnull().sum().sum())

data['Protocol'] = data['Protocol'].fillna(data['Protocol'].median())
print("Protocol Missing Values: ", data['Protocol'].isnull().sum().sum())

# data['Inbound'] = data['Inbound'].fillna(data['Inbound'].median())
# print("Innbound Missing Values: ", data['Inbound'].isnull().sum().sum())
```

```python
# --- Handling the Infinite Values ---

data.replace([np.inf, -np.inf], np.nan, inplace=True)
data.fillna(data.median(numeric_only=True), inplace=True)
```

```python
data.isnull().sum().sum()
print(np.isinf(data.select_dtypes(include=[float, int])).sum().sum())
```

#### Converting the Label values Benign(Normal non-attack state) -> 0 and other attacks -> 1 

```python
data['Label'] = data['Label'].apply(lambda x: 0 if x == 'BENIGN' else 1)
```
